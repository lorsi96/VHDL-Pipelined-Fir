```python
import matplotlib.pyplot as plt
import numpy as np

data = np.load('../output/result_sine_filtered_cocotb.npy')
data2 = np.load('../output/result_sine_filtered_numpy.npy')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(30, 10))
ax[0].plot(data[:len(data2)] - data2)
ax[0].set_xlabel('Samples [N]')
ax[0].set_ylabel('Error')
ax[0].set_title('Error: Seal numpy - Filtro FIR PDS')
ax[1].legend('Error')
ax[0].grid()
ax[1].plot(data2)
ax[1].plot(data[:len(data2)])
ax[1].set_xlabel('Samples [N]')
ax[1].set_ylabel('Magnitud')
ax[1].legend(['Numpy', 'CocoTB'])
ax[1].set_title('Seal numpy vs Seal DUT CocoTb')
ax[1].grid()
```
